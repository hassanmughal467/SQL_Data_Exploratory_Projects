```python
import sys
import pypyodbc as odbc
import requests
import pandas as pd
import json
from prettytable import PrettyTable
```

```python
tableobj=PrettyTable()
keyVal='fd7524aa-31c7-41dc-9cb2-16acfdbbdc48'
api_endpoint="https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest?CMC_PRO_API_KEY="
api_endpoint+=keyVal
json_Data=requests.get(api_endpoint).json()
cryptodata=json_Data['data']
```

```python
records=list()
```

```python
for currency in cryptodata:
    currency_name=currency['name']
    currency_price=currency['quote']['USD']['price']
    currency_change_1h=currency['quote']['USD']['percent_change_1h']
    currency_change_24h=currency['quote']['USD']['percent_change_1h']
    currency_change_7d=currency['quote']['USD']['percent_change_1h']
    currency_change_30d=currency['quote']['USD']['percent_change_1h']
    st=str(currency_name)+"|"+str(currency_price)+"|"+str(currency_change_1h)+"|"+str(currency_change_24h)+"|"+str(currency_change_7d)+"|"+str(currency_change_30d)
    p=st.split("|")
    records.append(p)
```

```python
records
```

```python
DRIVER = 'SQL Server'
SERVER_NAME = 'DESKTOP-R7DOSJA'
DATABASE_NAME = 'Cryptocurrency'

conn_string = f"""
    Driver={{{DRIVER}}};
    Server={SERVER_NAME};
    Database={DATABASE_NAME};
    Trust_Connection=yes;
"""

try:
    conn = odbc.connect(conn_string)
except Exception as e:
    print(e)
    print('task is terminated')
    sys.exit()
else:
    cursor = conn.cursor()


insert_statement = """
    INSERT INTO crypto
    VALUES (?, ?, ?, ?,?,?)
"""

try:
    for record in records:
        print(record)
        cursor.execute(insert_statement, record)        
except Exception as e:
    cursor.rollback()
    print(e.value)
    print('transaction rolled back')
else:
    print('records inserted successfully')
    cursor.commit()
    cursor.close()
finally:
    if conn.connected == 1:
        print('connection closed')
        conn.close()
```
